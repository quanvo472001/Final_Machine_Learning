```python
import tensorflow as tf
from keras.layers import Dense, LSTM, RNN,SimpleRNNCell
# Create window to create data
class WindowGenerator():
    def __init__(self, input_width, label_width, shift,
               train_df=train_df, val_df=val_df, test_df=test_df,
               label_columns=None):
        # Store the raw data.
        self.train_df = train_df
        self.val_df = val_df
        self.test_df = test_df

        # Work out the label column indices.
        self.label_columns = label_columns
        if label_columns is not None:
            self.label_columns_indices = {name: i for i, name in
                                        enumerate(label_columns)}
        self.column_indices = {name: i for i, name in
                               enumerate(train_df.columns)}

        # Work out the window parameters.
        self.input_width = input_width # input width
        self.label_width = label_width # output with
        self.shift = shift # degree of displacement

        self.total_window_size = input_width + shift

        self.input_slice = slice(0, input_width)
        self.input_indices = np.arange(self.total_window_size)[self.input_slice]

        self.label_start = self.total_window_size - self.label_width
        self.labels_slice = slice(self.label_start, None)
        self.label_indices = np.arange(self.total_window_size)[self.labels_slice]

    def __repr__(self):
        return '\n'.join([
            f'Total window size: {self.total_window_size}',
            f'Input indices: {self.input_indices}',
            f'Label indices: {self.label_indices}',
            f'Label column name(s): {self.label_columns}'])
```

Use the window to use 15 days of data and predict the next 5 days

```python
OUT_STEPS = 5
input_width =15
multi_window = WindowGenerator(input_width=input_width,
                               label_width=OUT_STEPS,
                               shift=OUT_STEPS,label_columns = ['Close_AAPL'])
# chọn label là giá đóng cửa, tất cả feature còn lại là input
multi_window
```

```python
#Get data from above indexs
def split_window(self, features):
    inputs = features[:, self.input_slice, :]
    labels = features[:, self.labels_slice, :]
    if self.label_columns is not None:
        labels = tf.stack(
            [labels[:, :, self.column_indices[name]] for name in self.label_columns],
            axis=-1)

  # Slicing doesn't preserve static shape information, so set the shapes
  # manually. This way the `tf.data.Datasets` are easier to inspect.
    inputs.set_shape([None, self.input_width, None])
    labels.set_shape([None, self.label_width, None])

    return inputs, labels

WindowGenerator.split_window = split_window
```

```python
def plot(self, model=None, plot_col='Close_AAPL', max_subplots=3):
    plt.figure(figsize=(20, 10))
    plot_col_index = self.column_indices[plot_col]
    max_n = min(max_subplots, len(inputs))
    # Draw input
    for n in range(max_n):
        plt.subplot(max_n, 1, n+1)
        plt.ylabel(f'{plot_col} [normed]')
        plt.plot(self.input_indices, inputs[n, :, plot_col_index],
                 label='Inputs', marker='.', zorder=-10)

        if self.label_columns:
            label_col_index = self.label_columns_indices.get(plot_col, None)
        else:
            label_col_index = plot_col_index

        if label_col_index is None:
            continue

        plt.scatter(self.label_indices, labels[n, :, label_col_index],
                    edgecolors='k', label='Labels', c='#2ca02c', s=64)
        # Draw predictions
        if model is not None:
            predictions = model(inputs)
            plt.scatter(self.label_indices, predictions[n, :, label_col_index],
                  marker='X', edgecolors='k', label='Predictions',
                  c='#ff7f0e', s=64)

        if n == 0:
            plt.legend()

    plt.xlabel('Time [day]')

WindowGenerator.plot = plot
```

```python
def make_dataset(self, data): #data lấy liên tục
    data = np.array(data, dtype=np.float32)
    ds = tf.keras.utils.timeseries_dataset_from_array(
      data=data,
      targets=None,
      sequence_length=self.total_window_size,
      sequence_stride=1,
      shuffle=True,#to avoid sequence bias
      batch_size=64,)

    ds = ds.map(self.split_window)

    return ds

WindowGenerator.make_dataset = make_dataset
```

```python
@property
def train(self):
    return self.make_dataset(self.train_df)

@property
def val(self):
    return self.make_dataset(self.val_df)

@property
def test(self):
    return self.make_dataset(self.test_df)

WindowGenerator.train = train
WindowGenerator.val = val
WindowGenerator.test = test
```

```python
for inputs, labels in multi_window.train.take(1):
    print(f'Inputs shape (batch, time, features): {inputs.shape}')
    print(f'Labels shape (batch, time, features): {labels.shape}')
```

```python
tf.keras.backend.clear_session()
MAX_EPOCHS = 30

# cho learning rate giảm dần theo epoch
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)
# train model
def compile_and_fit(model, window, patience=20):
    # Dừng việc train sớm nếu thấy tốt
    early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                                                    patience=patience,
                                                    mode='min')
    # Set the learning rate scheduler
    lr_schedule = tf.keras.callbacks.LearningRateScheduler(scheduler)


    model.compile(#loss=tf.keras.losses.MeanSquaredError()
                  loss=tf.keras.losses.Huber(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=[tf.keras.metrics.MeanAbsoluteError(),
                         tf.keras.metrics.RootMeanSquaredError()])

    history = model.fit(window.train, epochs=MAX_EPOCHS,
                      validation_data=window.val,
                      callbacks=[early_stopping,lr_schedule],verbose=1)
    return history
```

```python
def plot_train (history):
    # list all data in history
    print(history.history.keys())

    # summarize history for loss

    fig, ax = plt.subplots()
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'valid'], loc='upper left')
    plt.show()
```

We evaluate the model through the two best methods to evaluate a quantitative data model:

Root Mean Square Error (RMSE)
Mean absolute Error (MAE)

## Reccurent Neural Network

Build RNN model

```python
def build_RNN_Dense_model():

    input = tf.keras.Input(shape= (input_width,num_features))

    x = tf.keras.layers.SimpleRNN(30, return_sequences=True)(input)

    x = tf.keras.layers.SimpleRNN(30)(x)

    x = tf.keras.layers.Dense(OUT_STEPS*num_features)(x)

    output = tf.keras.layers.Reshape([OUT_STEPS, num_features])(x)

    model = tf.keras.Model(inputs=input, outputs=output)

    return model

RNN_Dense_model = build_RNN_Dense_model()
RNN_Dense_model.summary()
```

```python
import IPython
import IPython.display
history_RNN_Dense = compile_and_fit(RNN_Dense_model, multi_window)

multi_val_performance = {}
multi_performance = {}
IPython.display.clear_output()
multi_val_performance['RNN_Dense'] = RNN_Dense_model.evaluate(multi_window.val)
multi_performance['RNN_Dense'] = RNN_Dense_model.evaluate(multi_window.test, verbose=0)
plot_train(history_RNN_Dense)
multi_window.plot(RNN_Dense_model)
```

## Feed Forward Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
def build_FFNN_Dense_model():

    input = tf.keras.Input(shape= (input_width,num_features))

    x = tf.keras.layers.Flatten()(input)

    x = (tf.keras.layers.Dense(units=64, activation='relu')(x))

    x=(tf.keras.layers.Dense(units=32, activation='relu')(x))

    x=(tf.keras.layers.Dense(units=16, activation='relu')(x))

    x=(tf.keras.layers.Dense(OUT_STEPS*num_features)(x))

    output = tf.keras.layers.Reshape([OUT_STEPS, num_features])(x)
    model = Sequential()
    model = tf.keras.Model(inputs=input, outputs=output)


    return model
FFNN_Dense_model = build_FFNN_Dense_model()
FFNN_Dense_model.summary()
```

```python
history_FFNN_model = compile_and_fit(FFNN_Dense_model, multi_window)
IPython.display.clear_output()
multi_val_performance['FFNN_model'] = FFNN_Dense_model.evaluate(multi_window.val)
multi_performance['FFNN_model'] = FFNN_Dense_model.evaluate(multi_window.test, verbose=0)
plot_train(history_FFNN_model)
multi_window.plot(FFNN_Dense_model)
```
